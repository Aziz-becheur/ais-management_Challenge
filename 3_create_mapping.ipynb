# EQ-Klasse seem to be systematically very close to the anlagen-id where very close ids usually represent the same category

```python
import json
customer_structure = "./customer_data/structure.json"
#  Collect all categories from customer data
customer_strs = []
with open(customer_structure, 'r') as f:
    customer_data = json.load(f)

for cat in customer_data.keys():
    cat_str = cat
    for subcat in customer_data[cat]['subcategories']:
        for subcats in subcat.values():
            cat_str += f": {list(subcats.keys())[0]}: {list(subcats.values())[0]}"
    customer_strs.append(cat_str)

customer_strs
```

```python
reference_structure = "./reference_data/structure.json"
#  Collect all categories from reference data
reference_strs = []
with open(reference_structure, 'r') as f:
    reference_data = json.load(f)

for cat in reference_data.keys():
    cat_str = cat
    for subcat in reference_data[cat]['subcategories']:
        for subcat_name, subcat_values in subcat.items():
            cat_str += f": {subcat_name}: {subcat_values}"
    reference_strs.append(cat_str)

reference_strs
```

```python
print(f"referencecategories + subcategories: {reference_strs}")
print(f"customercategories + subcategories: {customer_strs}")
```

```python
reference_categories = list(reference_data.keys())
customer_categories = list(customer_data.keys())
reference_categories, customer_categories
```

# There is a huge difference between the customer categories and Reference categories
## A mapping is needed between both

```python
import torch
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer('paraphrase-multilingual-MiniLM-L12-v2')
customer_embeddings = model.encode(customer_categories, convert_to_tensor=True)
reference_embeddings = model.encode(reference_categories, convert_to_tensor=True)

for idx, cust_cat in enumerate(customer_categories):
    similarities = util.cos_sim(customer_embeddings[idx], reference_embeddings)[0]
    topk = torch.topk(similarities, 3)
    print(f"\n{cust_cat}:")
    for rank, (i, score) in enumerate(zip(topk.indices, topk.values)):
        print(f"  {rank+1}. {reference_categories[i]} (score: {score:.2f})")
```

# Very low accuracy, add subcategories to add more context

```python
customer_embeddings = model.encode(customer_strs, convert_to_tensor=True)
reference_embeddings = model.encode(reference_strs, convert_to_tensor=True)
scores_sim_l12 = {}
for idx, cust_cat in enumerate(customer_strs):
    similarities = util.cos_sim(customer_embeddings[idx], reference_embeddings)[0]
    topk = torch.topk(similarities, 5)
    print(f"\n{cust_cat.split(':')[0]}:")
    scores_sim_l12[cust_cat.split(' ;')[0]] = []
    for rank, (i, score) in enumerate(zip(topk.indices, topk.values)):
        print(f"  {rank+1}. {reference_strs[i].split(':')[0]} (score: {score:.2f})")
        scores_sim_l12[cust_cat.split(' ;')[0]].append((reference_strs[i].split(':')[0], score.item()))
```

## Try with Embeddings from OpenAI

```python
# try out openai embeddings for better results
def openai_embeddings(text):
    from openai import OpenAI
    client = OpenAI(api_key="your_openai_api_key_here")  # Replace with your OpenAI API key
    try:
        response = client.embeddings.create(
            input=text,
            model="text-embedding-3-small",
        )
        return response.data[0].embedding
    except:
        half_text = text[:len(text)//2]
        response = client.embeddings.create(
            input=half_text,
            model="text-embedding-3-small",
        )
        return response.data[0].embedding

customer_embeddings = [openai_embeddings(cat) for cat in customer_strs]
reference_embeddings = [openai_embeddings(cat) for cat in reference_strs]
customer_embeddings = torch.tensor(customer_embeddings)
reference_embeddings = torch.tensor(reference_embeddings)
scores_sim_openai = {}
for idx, cust_cat in enumerate(customer_strs):
    similarities = util.cos_sim(customer_embeddings[idx], reference_embeddings)[0]
    topk = torch.topk(similarities, 5)
    print(f"\n{cust_cat.split(':')[0]}:")
    scores_sim_openai[cust_cat.split(' ;')[0]] = []
    for rank, (i, score) in enumerate(zip(topk.indices, topk.values)):
        print(f"  {rank+1}. {reference_strs[i].split(':')[0]} (score: {score:.2f})")
        scores_sim_openai[cust_cat.split(' ;')[0]].append((reference_strs[i].split(':')[0], score.item()))
```

```python
assert scores_sim_l12.keys() == scores_sim_openai.keys()
# find a common ground between the two scores, give more weight to the openai scores, give best 3 only
scores_combined = {}
for key in scores_sim_l12.keys():
    combined_scores = {}
    for (ref_cat, score) in scores_sim_l12[key]:
        combined_scores[ref_cat] = score * 0.4  # weight for L12 scores
    for (ref_cat, score) in scores_sim_openai[key]:
        if ref_cat in combined_scores:
            combined_scores[ref_cat] += score * 0.6  # weight for OpenAI scores
        else:
            combined_scores[ref_cat] = score * 0.6  # only OpenAI score
    topk = sorted(combined_scores.items(), key=lambda x: x[1], reverse=True)[:3]
    scores_combined[key] = topk

scores_combined
```

# We can deduct a mapping between the categories now

```python
# Create a final mapping between customer categories and reference categories
final_mapping = {}
used_ref_cats = set()
for cust_cat, scores in scores_combined.items():
    final_mapping[cust_cat.split(':')[0]] = []
    for i,(ref_cat, score) in enumerate(scores):
        if score > 0.7 and i==0:
            final_mapping[cust_cat.split(':')[0]].append(ref_cat)
            used_ref_cats.add(ref_cat)
            break
        else:
            final_mapping[cust_cat.split(':')[0]].append(ref_cat)



print("Final Mapping between Customer Categories and Reference Categories:")
for cust_cat, ref_cats in final_mapping.items():
    print(f"{cust_cat.split(':')[0]}: {', '.join(ref_cats)}")
```

```python
# Final manual adjustments
final_mapping['RLT'] = ['Lufttechnische Anlagen']
final_mapping['Haustechn Einricht'] = ['Neu Ausstattung', 'Baukonstruktionen in Außenanlagen']
final_mapping['Brandschutz'] = ['Nutzungsspezifische Anlagen']
final_mapping['Türen und Tore'] = ['Außenwände', 'Innenwände', 'Dächer']
final_mapping['BuES, GLT, IZS'] = ['Gebäudeautomation']
final_mapping
```

```python
# save final mapping as JSON
with open('./outputs/mapping.json', 'w') as f:
    json.dump(final_mapping, f, indent=4)
print("Final mapping saved to final_mapping.json")
```

```python
#update structure of customer data
struct_file = "./customer_data/structure.json"
for map in final_mapping.keys():
    mapped = []
    for mapped_cat in final_mapping[map]:
        if mapped_cat in reference_data.keys():
            mapped.append(reference_data[mapped_cat]['id'])
    customer_data[map]['id'] = mapped
            
# save updated customer structure
with open("./outputs/mapped_structure.json", 'w') as f:
    json.dump(customer_data, f, indent=4)
```

# Now we do the mapping of each subcategory

### mapping of lufttechnische anlagen with RTL

```python
mapped_customer_data = customer_data.copy()
scores_sim_openai = {}
for cust_cat,ref in final_mapping.items():
    print(f"Mapping {cust_cat} subs with {ref} subs")
    reference_subcats = []
    customer_subcats = []
    for ref_cat in ref:
        for subcat in reference_data[ref_cat]['subcategories']:
            reference_subcats.append(f"{subcat}")
        for subcat in mapped_customer_data[cust_cat]['subcategories']:
            customer_subcats.append(f"{subcat}")

    # embed each item of the list and find the closest match of each sub cat from RLT to Lufttechnische Anlagen using openai embeddings
    reference_subcats_embeddings = [openai_embeddings(cat) for cat in reference_subcats]
    customer_subcats_embeddings = [openai_embeddings(cat) for cat in customer_subcats]
    cust_subcats_embeddings = torch.tensor(customer_subcats_embeddings)
    reference_subcats_embeddings = torch.tensor(reference_subcats_embeddings)
    for idx, cust_subcat in enumerate(customer_subcats):
        similarities = util.cos_sim(cust_subcats_embeddings[idx], reference_subcats_embeddings)[0]
        k = min(5, len(similarities))
        topk = torch.topk(similarities, k)
        print(f"\n{cust_subcat.split(':')[0]}:")
        scores_sim_openai[cust_subcat.split(' ;')[0].split(':')[0].split("'")[1]] = []
        for rank, (i, score) in enumerate(zip(topk.indices, topk.values)):
            print(f"  {rank+1}. {reference_subcats[i].split(':')[0]} (score: {score:.2f})")
            scores_sim_openai[cust_subcat.split(' ;')[0].split(':')[0].split("'")[1]].append((reference_subcats[i].split(':')[0].split("'")[1], score.item()))
```

```python
mapping_subs = {}
for score in scores_sim_openai.keys():
    mapping_subs[score] = scores_sim_openai[score][0][0]  # take only the first match for each customer subcategory

mapping_subs
```
